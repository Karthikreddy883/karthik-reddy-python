```python

'''
ord()--converts any character into ascii value
chr()---converts ascii (number) into a character
'''
for ch in range(ord('A'),ord('Z')+1):
  print(chr(ch))
for ch in range(ord('a'),ord('z')+1):
  print(chr(ch))
```
