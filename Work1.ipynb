```python
# problem statement 1
class Passenger:
    def __init__(self, name):
        self.name = name
class Flight:
    def __init__(self, flight_no):
        self.flight_no = flight_no
class BoardingPass:
    def __init__(self, passenger, flight, seat):
        self.passenger = passenger
        self.flight = flight
        self.seat = seat
    def generate_boarding_pass(self):
        print("=" * 50)
        print("               BOARDING PASS")
        print("=" * 50)
        print("\nPassenger Name :", self.passenger.name)
        print("Flight Number  :", self.flight.flight_no)
        print("Seat Number    :", self.seat)
        print("\nStatus         : CHECK-IN COMPLETE")
        print("\n" + "=" * 50)
class Airport:
    def __init__(self):
        self.passengers = []      # List
        self.occupied_seats = set()  # Set
    def register_passenger(self, passenger):
        self.passengers.append(passenger)
    def assign_seat(self, seat):
        if seat not in self.occupied_seats:
            self.occupied_seats.add(seat)
            return True
        else:
            return False
airport = Airport()
passenger = Passenger("Mason")
flight = Flight("AI202")
seat = "12A"
airport.register_passenger(passenger)
if airport.assign_seat(seat):
    boarding = BoardingPass(passenger, flight, seat)
    boarding.generate_boarding_pass()
else:
    print("Seat already occupied")
```

```python
# problem statement 2
class Vehicle:
    def __init__(self, number):
        self.number = number
class ParkingManager:
    def __init__(self, vehicle, hours, rate):
        self.vehicle = vehicle
        self.hours = hours
        self.rate = rate
    def calculate_fee(self):
        return self.hours * self.rate
    def generate_receipt(self):
        print("="*50)
        print("PARKING RECEIPT")
        print("="*50)
        print("Vehicle Number :", self.vehicle.number)
        print("Hours Parked   :", self.hours)
        print("Parking Fee    : $", self.calculate_fee())
vehicle = Vehicle("AP39AB1234")
parking = ParkingManager(vehicle, 5, 30)
parking.generate_receipt()
```

```python
# problem statement 3
class Player:
    def __init__(self, name):
        self.name = name
class Tournament:
    def __init__(self, player, runs):
        self.player = player
        self.runs = runs
    def report(self):
        total = sum(self.runs)
        matches = len(self.runs)
        avg = total / matches
        print("="*50)
        print("PLAYER PERFORMANCE REPORT")
        print("="*50)
        print("Player Name    :", self.player.name)
        print("Total Runs     :", total)
        print("Matches Played :", matches)
        print("Average Runs   :", avg)
player = Player("Arjun")
tournament = Tournament(player, [50, 75, 100])
tournament.report()
```

```python
# problem statement 4
class Guest:
    def __init__(self, name):
        self.name = name
class Room:
    def __init__(self, room_type, rate):
        self.room_type = room_type
        self.rate = rate
class Reservation:
    def __init__(self, guest, room, nights):
        self.guest = guest
        self.room = room
        self.nights = nights
    def bill(self):
        total = self.room.rate * self.nights
        print("="*50)
        print("HOTEL INVOICE")
        print("="*50)
        print("Guest Name    :", self.guest.name)
        print("Room Type     :", self.room.room_type)
        print("Nights Stayed :", self.nights)
        print("Total Amount  : $", total)
guest = Guest("Sophia")
room = Room("Deluxe", 3500)
Reservation(guest, room, 3).bill()
```

```python
# problem statement 5
class Customer:
    def __init__(self, name, mobile):
        self.name = name
        self.mobile = mobile
class RechargePlan:
    def __init__(self, plan, amount):
        self.plan = plan
        self.amount = amount
class Recharge:
    def __init__(self, customer, plan):
        self.customer = customer
        self.plan = plan
    def generate_receipt(self):
        print("="*50)
        print("RECHARGE RECEIPT")
        print("="*50)
        print("Customer Name :", self.customer.name)
        print("Plan Selected :", self.plan.plan)
        print("Amount Paid   : $", self.plan.amount)
        print("Status        : SUCCESSFUL")

cust = Customer("Logan", "9876543210")
plan = RechargePlan("Unlimited 84 Days", 799)
Recharge(cust, plan).generate_receipt()
```

```python
# problem statement 6
class Client:
    def __init__(self, name):
        self.name = name
class Project:
    def __init__(self, name, hours, rate):
        self.name = name
        self.hours = hours
        self.rate = rate
class Invoice:
    def __init__(self, client, project):
        self.client = client
        self.project = project
    def generate(self):
        total = self.project.hours * self.project.rate
        print("="*50)
        print("CLIENT INVOICE")
        print("="*50)
        print("Client Name  :", self.client.name)
        print("Project Name :", self.project.name)
        print("Hours Worked :", self.project.hours)
        print("Total Amount : $", total)

client = Client("Olivia")
project = Project("Portfolio Website", 20, 1000)
Invoice(client, project).generate()
```

```python
# problem statement 8
class Player:
    def __init__(self, name):
        self.name = name
class Tournament:
    def __init__(self, player, scores):
        self.player = player
        self.scores = scores
    def report(self):
        total = sum(self.scores)
        print("="*50)
        print("TOURNAMENT REPORT")
        print("="*50)
        print("Player Name :", self.player.name)
        print("Final Score :", total)
        if total >= 400:
            print("Rank Status : QUALIFIED")
        else:
            print("Rank Status : NOT QUALIFIED")
player = Player("Leo")
Tournament(player, [100, 150, 200]).report()
```

```python
# problem statement 9
class Member:
    def __init__(self, name):
        self.name = name
class MembershipPlan:
    def __init__(self, plan, fee):
        self.plan = plan
        self.fee = fee
class Gym:
    def __init__(self, member, plan, duration):
        self.member = member
        self.plan = plan
        self.duration = duration
    def receipt(self):
        total = self.plan.fee * self.duration
        print("="*50)
        print("MEMBERSHIP RECEIPT")
        print("="*50)
        print("Member Name :", self.member.name)
        print("Plan        :", self.plan.plan)
        print("Duration    :", self.duration, "Months")
        print("Total Fee   : $", total)

member = Member("Ava")
plan = MembershipPlan("Premium", 2000)
Gym(member, plan, 6).receipt()
```

```python
# problem statement 10
class User:
    def __init__(self, name):
        self.name = name
class Wallet:
    def __init__(self, balance):
        self.balance = balance
        self.transactions = []
    def transfer_money(self, amount):
        self.balance -= amount
        self.transactions.append(amount)
class Transaction:
    def __init__(self, user, wallet):
        self.user = user
        self.wallet = wallet
    def receipt(self, opening, transfer):
        print("="*50)
        print("TRANSACTION RECEIPT")
        print("="*50)
        print("User Name       :", self.user.name)
        print("Opening Balance : $", opening)
        print("Transfer Amount : $", transfer)
        print("Current Balance : $", self.wallet.balance)
        print("Status          : SUCCESSFUL")
user = User("Noah")
wallet = Wallet(10000)
opening_balance = wallet.balance
wallet.transfer_money(2500)

Transaction(user, wallet).receipt(opening_balance, 2500)
```
