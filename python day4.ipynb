```python
size = input("Enter the size of pizza: ")
if size == "small":
    bill = 10
elif size == "medium":
    bill = 15
elif size == "large":
    bill = 20
else:
    print("Invalid pizza size")
    exit()
print(f"available toppings are \n 1.cheese\n 2.pepperoni\n 3.cheeseolives\n 4.jalapanos")
n = int(input("Enter the number of toppings: "))
for i in range(n):
    topping = input("Enter your topping: ")
    if topping == "cheese":
        bill += 2
    elif topping == "pepperoni":
        bill += 3
    elif topping == "olives":
        bill += 5
    elif topping == "jalapanos":
        bill += 5
    else:
        print("Invalid topping")
        exit()
print(f"Total bill is:${bill}")
```

```python
#shallow copy
import copy
original=[1,2,3,4,5]
print(original)
new=original
print(new)
new[0]=100
print(original)
print(new)
```

```python
#deep copy
import copy
original=[1,2,3,4,5]
print(original)
new=copy.deepcopy(original)
print(new)
new[0]=100
print(original)
print(new)
```
