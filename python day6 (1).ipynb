```python
# Initial stock
stock = {'P001': 100, 'P002': 50}

# Transactions
transactions = [('P001', 20, 'OUT'),('P002', 30, 'IN')]
for pid, qty, typ in transactions:

    # If product not in stock, initialize with 0
    if pid not in stock:
        stock[pid] = 0

    # Incoming stock
    if typ == 'IN':
        stock[pid] += qty

    # Outgoing stock
    elif typ == 'OUT':
        if stock[pid] >= qty:
            stock[pid] -= qty
        else:
            print("Insufficient Stock for", pid)

# Display updated stock
for pid, qty in stock.items():
    print(pid + ":", qty)
```
